```python
import pandas as pd
import numpy as np
```

```python
#Loading the dataset of the case.

df = pd.read_excel('/content/house (1).xlsx')
```

```python
df.describe()
```

```python
df
```

```python
#Verifying the columns of the object type.

df.info()
```

```python
#Removing the NaN values

df = df.dropna()
```

```python
#Counting the values os the objects variables to identify how to proceed.
print(df.ExterQual.value_counts())
print(df.ExterCond.value_counts())
```

# Dummies #


```python
# We can use the get dummies

df = pd.get_dummies(df)
df.head()
```

# Treinando #

```python
y = np.array(df['SalePrice'])

x =  df.drop("SalePrice", axis = 1)
```

```python
x_list = list(x.columns)
```

```python
x_list
```

```python
x = np.array(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
from sklearn.datasets import make_moons
from sklearn import metrics
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
```

```python
# Training our base of train

rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
rf.fit(train_x, train_y)
```

```python
# Training the basetest
predction_rf= rf.predict(test_x)

error_rf = abs(predction_rf - test_y)

r_square = rf.score(x, y)

print('error', error_rf)
print('R^2', r_square)
print('MAE', metrics.mean_absolute_error(test_y, predction_rf))
print('MSE', metrics.mean_squared_error(test_y, predction_rf))
```

```python
ada = AdaBoostRegressor(n_estimators = 1000)
ada.fit(train_x, train_y)

ada_pred =  ada.predict(test_x)
```

```python
error_ada = abs(ada_pred - test_y)

r_square = ada.score(x, y)

print('error', error_ada)
print('R^2', r_square)
print('MAE', metrics.mean_absolute_error(test_y, ada_pred))
print('MSE', metrics.mean_squared_error(test_y, ada_pred))
```

```python
gbr = GradientBoostingRegressor(n_estimators = 100)
gbr.fit(train_x, train_y)

gbr_pred =  gbr.predict(test_x)
```

```python
error_gbr = abs(gbr_pred - test_y)

r_square = gbr.score(x, y)

print('error', error_gbr)
print('R^2', r_square)
print('MAE', metrics.mean_absolute_error(test_y, gbr_pred))
print('MSE', metrics.mean_squared_error(test_y, gbr_pred))
```
